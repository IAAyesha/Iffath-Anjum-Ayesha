```python
books=book.iloc[:,1:]
books.head()
```

```python
#number of unique books in dataset
len(books['Book.Title'].unique())
```

```python
#converting long data into wide table using pivot_table
books_df=books.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
books_df
```

```python
books_df.index=books['User.ID'].unique()
```

```python
books_df
```

```python
#impute NaN values with 0 values
books_df.fillna(0,inplace=True)
books_df
```

```python
#calculating Cosine_similarity between users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
sim_user= 1-pairwise_distances(books_df.values,metric='cosine')
```

```python
sim_user
```

```python
#dataframe
sim_user_df=pd.DataFrame(sim_user)
sim_user_df
```

```python
#set index & column names to user ids
sim_user_df.index=books['User.ID'].unique()
sim_user_df.column=books['User.ID'].unique()
sim_user_df
```

```python
sim_user_df.iloc[0:10,0:10]
```

```python
np.fill_diagonal(sim_user,0)
sim_user_df.iloc[0:10,0:10]
```

```python
#most similar user
sim_user_df.idxmax(axis=1)
```
